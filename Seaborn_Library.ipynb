. Seaborn
    Functionality :
        1. Built on top of Matplotlib for advanced visualizations.
        2. Provides beautiful default themes and statistical plotting functions.
        
    Importance for Machine Learning

    EDA: Enhances understanding of data by creating insightful plots like heatmaps, pair plots, and violin plots.
    
    Feature Selection: Visualizes relationships and correlations between features.

    Improved Aesthetics: Creates publication-quality visuals effortlessly.

Use Cases
    General Purpose:
    Statistical plots (e.g., boxplots, histograms).
    Machine Learning:
    Visualizing data distributions using KDE or histograms.
    Displaying correlation matrices to identify strong relationships.

   

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
```

Note : Seaborn has some built in dataset ( basic ML dataset )


```python
# Total bill vs tip Dataset 
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.tail()
```

```python
tips.describe()
```

```python
# Visualize the data 
sns.relplot(data= tips, x= 'total_bill', y = 'tip', col='time',hue= 'smoker', style='smoker',size='size')
```

```python
# Load iris dataset 

iris = sns.load_dataset('iris')
```

```python
iris.head()
```

Scatter plot


```python
sns.scatterplot(x='sepal_length', y = 'petal_length', hue='species',data= iris)
```

```python
sns.scatterplot(x='sepal_length', y = 'petal_width', hue='species',data= iris)
```
